```python
# Homework Problem 2
# a = 5/6
# find the congruent element for 5 * mul_inv(6)
inv_6 = pow(6, -1, 71)
frac = 5 * inv_6
congruent = frac % 71
print("multiplicative inverse of 6: ", inv_6)
print("actual value to find the congruence for: ", frac)
print("Congruent value: ", congruent)
```

```python
# Verification
# a = 5/6
# find the congruent element for mul_inv(2) + mul_inv(3) % 71
inv_2 = pow(2, -1, 71)
inv_3 = pow(3, -1, 71)
frac = inv_2 + inv_3
congruent = frac % 71
print("multiplicative inverse of 2: ", inv_2)
print("multiplicative inverse of 3: ", inv_3)
print("actual value to find the congruence for: ", frac)
print("Congruent value: ", congruent)
```
